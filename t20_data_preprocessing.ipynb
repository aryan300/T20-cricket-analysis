```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
```

# Match Results 

```python
with open('t20_json_files/t20_wc_match_results.json') as f: # Reading the json files
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary']) # Creating a dataframe
df_match.head()
```

```python
df_match.shape
```

```python
#Using Scorecard as a match id for linking with other tables by renaming scorecard to match_id
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True) #the name of the column is changed to match_id
df_match.head()
```

```python
#Create a match ids dictionary that maps team names to a unique match id.
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
```

# Batting Summary

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f: #Reading the batting_summary json file
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary']) 
  
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out") #Creating a new column out/not-out based on dismissal column
df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"], inplace=True) # The dismissal column is dropped
df_batting.head(10)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting.shape
```

```python
df_batting.to_csv('t20_csv_files/fact_bating_summary.csv', index = False)
```

# Bowling Summary

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```

# Players Information

```python
with open('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_bowling_mod = df_bowling.copy()

df_bowling_mod["economy"] = (
    df_bowling_mod["economy"]
    .astype(str)
    .str.extract(r"(\d+\.\d+)$")[0]
    .astype(float)
)
```

```python
plt.style.use("ggplot")

top_bowlers = (
    df_bowling_mod.dropna(subset=["economy"])
    .groupby("bowlerName")["economy"]
    .mean()
    .sort_values(ascending = False)
    .head(10)
)
plt.figure(figsize=(10,5))
top_bowlers.plot(kind="bar", color="lightgreen")
plt.title("Top 10 Bowlers by Economy Rate")
plt.ylabel("Economy")
plt.xlabel("Player")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
df_batting_mod = df_batting.copy()

df_batting_mod["SR"] = (
    df_batting_mod["SR"]
    .astype(str)
    .str.extract(r"(\d+\.\d+)$")[0]
    .astype(float)
)
```

```python
top_batsmen = (
    df_batting_mod[df_batting_mod["SR"].notna()]
    .groupby("batsmanName")["SR"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
top_batsmen.plot(kind="bar")
plt.title("Top 10 Batsmen by Strike Rate")
plt.ylabel("Strike Rate")
plt.xlabel("Player")
plt.xticks(rotation=45, ha="right")
plt.show()
```
